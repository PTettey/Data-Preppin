### Prepping Data Challenge: Un-pivoting (Week 18)
 

### Requirements
- Input the data
- Dynamically rename the fields so that there is a common separator between the Bike Type, Date and Measure Name
- Pivot the data
- Split out the Bike Type, Date and Measure Name
- Create a field for Sales and Profit
- Output the data

```python
import pandas as pd
import numpy as np
```

```python
# Input the data.
df = pd.read_csv("2022W18 Input.csv")
```

```python
df.head()
```

```python
#Dynamically rename the fields so that there is a common separator between the Bike Type, Date and Measure Name
df.columns
```

```python
df.columns = df.columns.str.replace('(___)',':')
```

```python
df.columns
```

```python
#Pivot the data
df = df.melt(id_vars = 'Region', var_name = 'other')
```

```python
#Split out the Bike Type, Date and Measure Name
df['Bike Type'] = df['other'].str.extract('(\w*):')
df['Month'] = df['other'].str.extract(':(.*):')
pd.to_datetime(df['Month'], format = '%b_%y')
df['sales/profit'] = df['other'].str.extract(':(\w*)$')
```

```python
df['Month'].unique()
```

```python
df['Bike Type'].unique()
```

```python
df['sales/profit'].unique()
```

```python
df['Month'] = pd.to_datetime(df['Month'], format = '%b_%y')
```

```python
df.head(10)
```

```python
#Create a field for Sales and Profit
df2 = df[["Bike Type","value","Region","Month","sales/profit"]]
df2['Sales'] = df.apply(lambda x : x['value'] if x['sales/profit'] == 'Sales' else None, axis=1)
df2.dropna(subset = ['Sales'], inplace=True)
```

```python
df2.head()
```

```python
df3 = df[["Bike Type","value","Region","Month","sales/profit"]]
df3['Profit'] = df3.apply(lambda x : x['value'] if x['sales/profit'] == 'Profit' else None, axis=1)
df3.dropna(subset = ['Profit'], inplace=True)
```

```python
df3.head()
```

```python
#Output the data
df = df2.merge(df3, on =["Bike Type","Region","Month"], how="inner")
```

```python
df = df[['Bike Type','Region','Month','Sales','Profit']]
```

```python
#output the data
df.to_csv('wk18-output.csv', index=False)
```
